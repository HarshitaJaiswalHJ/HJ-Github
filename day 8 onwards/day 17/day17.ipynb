```python
import matplotlib.pyplot as plt
```

#### Types of Graphs
1. plot--------------> line graph,between two parameters
2. bar---------------> bar graph of single parameter (optional catagorical data)
3. hist--------------> histogram of a bunch of data i.e frequency of single parameter
4. scatter-----------> dot graph between two parameters
5. boxplot/box-------> Statsical overview(mean,outliers,quantiles(25%,75%),IQR) in form of box for a single parameter
                       (optional-categorical data)
6. stem--------------> more like bar graph, lines instead of bars
7. violinplot--------> Violin shape statistical overview graph
                       spread of data (more like hist)
                       single parameter (optional catagorical data)

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
x1=np.arange(1,11,3)
```

```python
x1
```

```python
y=x**3
```

```python
plt.plot(x,y)
plt.title("plot Graph")
# plt.show()
```

```python
plt.bar(x,y)
plt.title("Bar Graph")
```

```python
plt.hist(y)
plt.title("hist Graph")
```

```python
plt.scatter(x,y)
plt.title("scatter Graph")
plt.grid()
plt.xlabel("x-->")
plt.ylabel("y-->")
plt.xticks(x)
plt.yticks(range(0,200,10))
```

```python
plt.boxplot(y)
plt.title("boxplot Graph")
```

```python
plt.stem(x,y)
plt.title("stem Graph")
```

```python
plt.violinplot(y)
plt.title("voilinplot Graph")
```

```python
y2=x**4
```

```python
plt.plot(x, y,x, y2)
plt.legend(['y=x**3','y2=x**4'])
plt.show()
```

```python
plt.plot(x,y,lw=3,marker='o',markersize=15,markerfacecolor='purple') 
plt.plot(x, y2,marker='v',markersize=10)
plt.legend(['y=x**3','y2=x**4'])
plt.show()
```

```python
fig=plt.figure(figsize=(10,5))
plt.plot(x,y,lw=3,marker='o',markersize=15,markerfacecolor='purple') 
plt.plot(x, y2,marker='v',markersize=10)
plt.legend(['y=x**3','y2=x**4'])
plt.grid()
plt.show()
```
