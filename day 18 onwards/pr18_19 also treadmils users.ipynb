### Graph Functions
1. plot------------------------------> simple Line plot
2. plot.area-------------------------> Filled version of line plot
3. plot.bar  /  plot.barth----------> Bar graph
4. plot.box-------------------------> Box plot
5. plot.density----------------------> Density line (smooth curved)
6. plot.hexbin-----------------------> Hexagonal version of scatter with Density
7. plot.hist-------------------------> Histogram
8. plot.kde--------------------------> Kernel Density Estimator
10. plot.pie-------------------------> Pie Chart of Categorical Data
11. plot.scatter---------------------> Scatter Graph

```python
for col in cat_cols:
    for ncol in num_cols:
        plt.figure(figsize=(10,3))
        plt.suptitle(f'{col} vs  {ncol}')
        plt.subplot(1,3,1)
        sns.boxplot(data=data, x=col, y=ncol, palette=choice(p_list))
        plt.subplot(1,3,2)
        sns.violinplot(data=data, x=col, y=ncol, palette=choice(p_list))
        plt.subplot(1,3,3)
        sns.barplot(data=data, x=col, y=ncol, palette=choice(p_list))
        plt.show()
```

```python
sns.pairplot(data=data[num_cols])
pg=sns.PairGrid(data=data[num_cols])
pg.map_diag(plt.hist)
pg.map_lower(sns.regplot)
pg.map_upper(sns.kdeplot)
```

3. Multivariate Analysis

```python
for col in cat_cols:
    if col=='Product':
        continue
    for ncol in num_cols:
        plt.figure(figsize=(10,3))
        plt.suptitle(f'{col} vs  {ncol}')
        plt.subplot(1,3,1)
        sns.boxplot(data=data, x=col, y=ncol, hue='Product', palette=choice(p_list))
        plt.subplot(1,3,2)
        sns.violinplot(data=data, x=col, y=ncol, hue='Product', palette=choice(p_list))
        plt.subplot(1,3,3)
        sns.barplot(data=data, x=col, y=ncol, hue='Product', palette=choice(p_list))
        plt.show()
```
