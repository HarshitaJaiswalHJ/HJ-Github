### Graph Functions
1. plot------------------------------> simple Line plot
2. plot.area-------------------------> Filled version of line plot
3. plot.bar  /  plot.barth----------> Bar graph
4. plot.box-------------------------> Box plot
5. plot.density----------------------> Density line (smooth curved)
6. plot.hexbin-----------------------> Hexagonal version of scatter with Density
7. plot.hist-------------------------> Histogram
8. plot.kde--------------------------> Kernel Density Estimator
10. plot.pie-------------------------> Pie Chart of Categorical Data
11. plot.scatter---------------------> Scatter Graph

```python
df2.plot()
```

```python
df2.plot(); plt.title("Line Plot"); plt.show()
 
df2.plot.area(); plt.title("Area Plot"); plt.show() 

df2.plot.bar(); plt.title("Bar Plot"); plt.show()
 
df2.plot.box(); plt.title("Box Plot"); plt.show()
 
df2.plot.barh(); plt.title("Bar Horizontal Plot"); plt.show()
 
df2.plot.density(); plt.title("Density Plot"); plt.show()
 
df2.plot.kde(); plt.title("Kde Plot"); plt.show()
 
df2['a'].plot.hist(); plt.title("Histogram Plot"); plt.show()
 
df2.plot.scatter(x='a',y='b'); plt.title("scatter Plot"); plt.show()

df2.plot.hexbin(x='a',y='b',C='c',gridsize=6); plt.title("Hexbin Plot"); plt.show()
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
from xml.etree.ElementInclude import include


tips.describe(include="all")
```

```python
tips["day"].value_counts()
```

```python
tips.groupby(["sex","time"])["day"].value_counts().sort_index()
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='day',data=tips,hue="sex")
```

```python
sns.barplot(x='day',y="total_bill",data=tips)
```

```python
sns.set_style("whitegrid")
```

```python
sns.barplot(x='day',y="total_bill",data=tips,estimator=sum)
```

```python
sns.barplot(x='day',y="total_bill",data=tips,estimator=np.mean,hue="time")
```

```python
sns.boxplot(x='day',y="total_bill",data=tips,hue="smoker")

```

```python
sns.violinplot(x='day',y="total_bill",data=tips,hue="smoker")
```

```python
sns.swarmplot(x='day',y="total_bill",data=tips,hue="smoker")
```

## DAY 19 Seaborn

```python
sns.distplot(tips['total_bill'])
```

```python
sns.displot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', y='tip')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip')
plt.show()
sns.jointplot(data=tips, x='total_bill', y='tip', hue='smoker')
plt.show()
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
plt.show()
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
fig= plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
plt.subplot(1,2,2)
sns.kdeplot(data=tips, x='total_bill', y='tip')
sns.rugplot(data=tips, x='total_bill', y='tip')
plt.show()
```
